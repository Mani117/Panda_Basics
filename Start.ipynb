```python
# Aggregating using DataFrame and map function
# In which axis=0 along the rows, and axis=1 along the columns
people = pd.DataFrame(np.random.randn(5, 5), columns=['a', 'b', 'c', 'd', 'e'], index=['Joe', 'Steve', 'Wes', 'Jim', 'Travis'])
print(people)
some_dict = {'a':"red", 'b':"red", 'c':"blue", 'd':"blue", 'e':"blue"}
sum_1 = people.groupby(some_dict, axis =1).sum()
print(sum_1)

# Using Series in pandas
series_1 = pd.Series(some_dict)
sum_2 = people.groupby(series_1, axis =1).sum()
print(sum_2)

# Sum using length of indexes
sum_2 = people.groupby(len).sum()
print(sum_2)
```

```python
# Data Aggregation
# Aggregations refer to any data transformation that produces scalar values from arrays.
#count:  Number of non-NA values in the group
#sum: Sum of non-NA values
#mean: Mean of non-NA values
#median: Arithmetic median of non-NA values
#std, var: Unbiased (n â€“ 1 denominator) standard deviation and variance
#min, max: Minimum and maximum of non-NA values
#prod: Product of non-NA values
#first, last: First and last non-NA values
#quantile: 

def peak_to_peak(arr):
    return arr.max() - arr.min()

group_ag = df2.groupby("key1")
peak_to_p = group_ag.agg(peak_to_peak)
print(peak_to_p)
group_ag.describe()

#Column Wise multiple function Application with function names

print(df7.head(10))
df7["tip_pct"] = df7['tip']/df7['total_bill']
print(df7.head(10))

grouped_ag = df7.groupby(['day', 'smoker'])
grouped_pct = grouped_ag["tip_pct"]

print(grouped_pct.agg('mean'))

# Finding multiple statistics of a given column of a data. It can also find statistics for all columns of data.
print(grouped_ag.agg(['mean','std',peak_to_peak]))

# Rename statistics head as you want
grouped_pct.agg([('foo', 'mean'), ('bar', np.std)])

# Example
grouped_stat_calc = grouped_ag.agg(['mean','std',peak_to_peak])
print(grouped_stat_calc['total_bill'])


# What if i want to calculate different statistics for different tracks
# We use dictionary for input where { column_name: function_to_be_applied}
grouped_ag.agg({'tip' : np.max, 'total_bill' : 'sum'})

# In case i want multiple statistcs for one column and some for other
grouped_ag.agg({'tip_pct' : ['min', 'max', 'mean', 'std'], 'size' : 'sum'})

# hierarchical Data removed
df7.groupby(['day', 'smoker'], as_index=False).mean()



# Sorting values by given column tip_pct
def top(df, n=5, column='tip_pct'):
    return df.sort_values(by=column)[:n]

top(df7, n = 5)

df7.groupby(['smoker', 'day']).apply(top, n=1, column='total_bill')
```
