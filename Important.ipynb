## Permutation and Random Sampling

```python
sampler = np.random.permutation(5)  # Creates 0-4 random no.s
# Suppose we have a dataframe
df = pd.DataFrame(np.arange(5 * 4).reshape((5, 4)))
# Now using iloc or take function we can permutate rows of above dataframe
df.iloc[sampler]
df.take(sampler)
# To select a random subset without replacement, you can use the sample method on Series and DataFrame
df.sample(n = 3)
# use replace = True for with replacement. Also obvious but to state, with replace = True, more than length
# of df can be extracted as it is with replacement
df.sample(n = 10, replace = True)
```
